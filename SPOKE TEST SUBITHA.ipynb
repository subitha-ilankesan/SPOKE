```python
#3. What does the distribution of waist sizes look like?
```

```python
waist_only = data[~data["product_type"].isin(['Gift Card','T-Shirts','Polo'])] # excluding products which do not have a waist size
waist_only.head()
```

```python
waist_only.loc[:,'waist_size_int'] = waist_only['variant_title'].str[:2].astype(int) #extracting the first 2 characters to determine waist size only, adding this as a column to the waist_only table
```

```python
waist_only
```

```python
waist_only.hist(column='waist_size_int')
```

```python
waist_only[['waist_size_int']].describe() 
```

```python
#3. What does the distribution of waist sizes look like? - see above
# Most customers have a waist size of 34. There is generally a normal distribution of waist sizes.
```

```python
#4. What is our unit return rate?
```

```python
excl_giftcard = data[~data["product_type"].isin(['Gift Card'])] #excluding gift cards from the data
```

```python
returns = excl_giftcard['returned_item_quantity'].abs() #turning negative values into positives for returns
returns.head()
```

```python
total_returns = returns.sum()
total_returns
```

```python
total_sales = excl_giftcard['ordered_item_quantity'].sum()
total_sales
```

```python
return_rate = total_returns/total_sales
return_rate * 100
```

```python
#4. What is our unit return rate?
# 18.3% This is below industry average (20-30% as reported by Shopify from GlobalData reports)
```

```python
#4b. What is our unit return rate for customers who ordered multiple sizes?
```

```python
multi = waist_only.groupby('customer_id').filter(lambda row: row.waist_size_int.nunique() > 1) #filtering the data by those who have ordered multiple sizes
```

```python
multi_returns = multi['returned_item_quantity'].abs()
tot_multi_returns = multi_returns.sum()
tot_multi_returns
```

```python
multi_sales = multi['ordered_item_quantity'].sum()
multi_sales
```

```python
multi_return_rate = tot_multi_returns/multi_sales
multi_return_rate * 100
```

```python
#4b. What is our unit return rate for customers who ordered multiple sizes?
# 22.7%, showing that those who order multiple sizes are more likely to return
```
